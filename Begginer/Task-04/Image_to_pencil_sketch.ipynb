**Importing Libraries**

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

**Read the image and show the original Image.**

```python
img = cv2.imread("home3.jpg")
cv2_imshow(img)
```

**Create a new Image by converting original image to greyscale**.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

**Invert the new grayscale image.**

```python
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

**Blur the Image using Gaussian function.**

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)

cv2_imshow(img_smoothing)
```

**Invert the blurred image**

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

**Pencil sketch of the given image.**

```python
final_img = dodgeV2(img_gray, img_smoothing)

cv2_imshow(final_img)
```
